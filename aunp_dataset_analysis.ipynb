```python
string_mats = []
material_mats = []
material_key_mats = []
    
for datum in data:
    for para in datum['paragraphs']:
        if para['contains_recipe']:
            for sent in para['sentences']:
                for mat in sent['all_materials']:
                    if type(mat) == str:
                        if para['_id'] not in string_mats:
                            string_mats.append(para['_id'])
                    elif 'material' not in mat.keys():
                        if para['_id'] not in material_key_mats:
                            material_key_mats.append(para['_id'])
                    elif mat['material'] == 'material':
                        if para['_id'] not in material_mats:
                            material_mats.append(para['_id'])
    print('{}/{}'.format(data.index(datum), len(data)), end='\r')
    
print('\n')
    
print(len(string_mats))
print(len(material_mats))
print(len(material_key_mats))
print(material_key_mats)
```

```python
def create_precursors_dict():
    precursors_dict = dict()

    for datum in data:
        for para in datum['paragraphs']:
            if para['contains_recipe']:
                for sent in para['sentences']:
                    del_idxs = []
                    for precursor in sent['precursors']:
                        #prec = list(precursor.keys())[0]
                        if type(precursor) != dict:
                            print(datum['doi'])
                            print(precursor)
                            print('\n')
                            del_idxs.append(sent['precursors'].index(precursor))
                            continue
                        elif 'material' not in precursor.keys():
                            del_idxs.append(sent['precursors'].index(precursor))
                            continue
                        else:
                            prec = precursor['material']
                        if prec in precursors_dict.keys():
                            if datum['doi'].lower() not in precursors_dict[prec]:
                                precursors_dict[prec].append(datum['doi'].lower())
                        else:
                            precursors_dict[prec] = [datum['doi'].lower()]
                    for correction, idx in enumerate(del_idxs):
                        del sent['precursors'][idx-correction]
    
    return precursors_dict
```

```python
aunp_precursors_dict = create_precursors_dict()

# clean precursors dict

with open('./rsc/aunp_precursor_syns_regex.json', 'r') as fp:
    prec_regex_map = json.load(fp)

with open('./rsc/aunp_garbage_precs.json', 'r' ) as fp:
    garbage_precs = json.load(fp)

# just check for unaccounted-for precursors
print('Unaccounted precursors: ')
print('========================')
for raw_prec in aunp_precursors_dict:
    contained = False
    for prec in prec_regex_map:
        combined_regex = "(" + ")|(".join(prec_regex_map[prec]) + ")"
        if re.match(combined_regex, raw_prec) or raw_prec in prec_syn_map.keys() or raw_prec in garbage_precs:
            contained = True
        if contained:
            break
    if contained:
        continue
    else:
        #print(raw_prec)
        continue
print('\n')

# take out the trash
for garb in garbage_precs:
    if garb in aunp_precursors_dict.keys():
        del aunp_precursors_dict[garb]

# consolidate synonyms
#for prec in prec_syn_map:
#    combined_regex = "(" + ")|(".join(prec_regex_map[prec]) + ")"
#    extend = True if prec in aunp_precursors_dict.keys() else False
#    aunp_precursors_dict = consolidate_syns(aunp_precursors_dict, prec, prec_syn_map[prec], extend=extend)

aunp_precursors_info = dict()
for key in prec_regex_map:
    aunp_precursors_info[key] = []

for raw_prec in aunp_precursors_dict:
    for key in prec_regex_map:
        combined_regex = "(" + ")|(".join(prec_regex_map[key]) + ")"
        if re.match(combined_regex, raw_prec) or raw_prec == key:
            aunp_precursors_info[key].extend(aunp_precursors_dict[raw_prec])
```

```python
def morphfilter(list1):
    """
    remove general components like 'NP' (also remove partial substring match)
    """
    exceptlist=['np','Np','NP',
                'Ag','Pt','Pd','Fe2O3','Fe3O4','SiO2', #other materials
                'particle','Particle','crystal','Crystal','structure','Structure','cluster','Cluster',
                'micro',
               ]
    goodcomps=[]
    for comp in list1:
        indic=1
        for key in exceptlist:
            if key in comp: 
                indic=0
                break
        if indic==1: 
            goodcomps.append(comp)
    return goodcomps


def regex_search(keyset,text,debug=False):
    """
    input: keyset (dict)  {cat: list of regex keywords}
    output: list of dict with span, text, and category as keys.
    """
    searchresults=[]
    for cat,keys in keyset.items():
        searchresult = [{'text':search.group(),'category':cat} for key in keys for search in re.finditer(key,text)]

        if len(searchresult)>=1:
            searchresults.append(searchresult[0])
            
    if len(searchresults)>1:
        if debug:
            print("multiple regex match in many categories")
            print(searchresults,text)
        return searchresults
    elif len(searchresults)==0:
        return 0
    elif len(searchresults):
        return searchresults
    
def morphnormalize(morphslist,regexinput=False): 
    """
    **this docstring is outdated**
    input: list of morphs in a paragraph
    output: list ( set of CATEGORIZED morphs )
    dependency: Morph_Dict
    * Also collect 'Other' morphologies in Other_Morph
    Error: return [] when multiple
    """       
    return list(set([ comp['category'] for item in morphslist if isinstance(item,list) for comp in item ]))
```

```python
syns_map = {
    'rod' : ['(Au ?|G)?NR\S*', 
             '(Au |[gG]old )?([nN]ano)?[rR]od\S*',          
            ],
    'sphere' : [
             '(Au |[gG]old )?([nN]ano)?[sS]pher(e|oid|ical)\S*',],
    'cube' : [
            '(Au |[gG]old )?([nN]ano)?[cC]ub(e|oid|ic)\S*', 
            '(Au |[gG]old )?([nN]ano)?[hH]exahedr(a|al|on)\S*', 
           ],
    'octahedra': ['(Au |[gG]old )?([nN]ano)?[oO]ctahedr(a|al|on)\S*',],
    'hexagon' : ['(Au ?|G)?([nN]ano)?[hH]exagon(al)?\S*'],
    'star' : ['(Au |[gG]old )?([nN]ano)?[sS]tar\S*',],
    'wire' : ['(Au ?|G)?NWs?','(Au |[gG]old )?([nN]ano)?[wW]ire\S*',],
    'triangle' : ['(Au ?|G)?([nN]ano)?[tT]riangle\S*'],
    'plate' : ['(Au ?|G)?([nN]ano)?[pP]late\S*'],
    'tube' : ['(Au ?|G)?([nN]ano)?[tT]ube\S*'],
    'prismatic' : ['(Au ?|G)?([nN]ano)?[pP]rism\S*'],
    'pyramid' : ['(Au ?|G)?([nN]ano)?([bB]i)?[pP]yramid\S*']
}
```

```python
### Morphologies line plot

year_morph_map = {}

morphs_map = {
    'sphere' : [],
    'rod' : [],
    'tube' : [],
    'cube' : [],
    'wire' : [],
    'star' : [],
    'plate' : [],
    'hexagon' : [],
    'prismatic' : [],
    'octahedra' : [],
    'pyramid' : [],
    'triangle' : []
}

for year in np.arange(1998, 2022):
    year_morph_map[year] = {}
    for morph in morphs_map.keys():
        year_morph_map[year][morph] = []
    
year_sorted_data = sorted(data, key=lambda k: k['publication_year'])

accounted_dois = []

single_count = 0
for datum in year_sorted_data:
    datum_char_info = {
        'MOR' : [],
        'DES' : [],
        'SED' : [],
        'SIZ' : [],
        'UNT' : [],
        'MES' : []
    }
    morph_count = []
    for para in datum['paragraphs']:
        if para['contains_characterization'] or para['contains_recipe']:
            morphs_par = morphfilter(list(set([mor for mor in para['morphological_information']['descriptors'] + para['morphological_information']['morphologies']])))
            morphs_par_norm=[regex_search(syns_map,comp) for comp in morphs_par]           
            morphs_category_par = morphnormalize(morphs_par_norm)
            morph_count.extend(morphs_category_par)
                        
                                
    if len(set(morph_count)) == 1:
        single_count += 1
        year_morph_map[datum['publication_year']][morph_count[0]].append(datum['doi'].lower())
    else:
        #print(set(morph_count))
        continue

print(single_count)
rod_nums = []
for year in year_morph_map:
    rod_nums.extend(year_morph_map[year]['rod'])
    
print(len(set(rod_nums)))
#pprint(year_morph_map)
```
